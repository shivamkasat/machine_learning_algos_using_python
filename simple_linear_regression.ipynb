```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 9.0)

data = pd.read_csv('/home/kasat/data.csv')
```

```python
# data.dropna(axis = 0, how = 'any', inplace = True)
X = data.iloc[: , 0]
y = data.iloc[: , 1]

training_count = len(X)

# plt.scatter(X, y)
```

```python
theta_0 = 0
theta_1 = 0

epochs = 1000;
learning_rate = 0.0005

y_predict = theta_1 * X + theta_0

plt.scatter(X, y)
```

```python
#updating theta_0 and theta_1
for i in range(epochs):
    up_theta_0 = theta_0 - (learning_rate / training_count) * sum(theta_0 + theta_1 * X - y)
    up_theta_1 = theta_1 - (learning_rate / training_count) * sum((theta_0 + theta_1 * X - y) * X)
    theta_0 = up_theta_0
    theta_1 = up_theta_1
    y_predict = theta_0 + theta_1 * X
    plt.scatter(X, y_predict)


plt.scatter(X, y, color = 'blue')
plt.plot([min(X), max(X)], [min(y_predict), max(y_predict)], color='black')  # regression lineplt.show()
plt.show()
```
