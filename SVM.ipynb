```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#from sklearn.metrics import r2_score
#v = x0.w0+x1.w1+b
def hyperplane(x,w,b,v):
    return (-w[0]*x-b+v) / w[1]


df= pd.read_excel("/home/kasat/datasets/Social_N.xlsx")

X = df.iloc[:,2:4]#not taking AGE feature in consideration
y = df.iloc[:,-1]



from sklearn.model_selection import train_test_split
X1,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25)
#plotting the dataset using make_blobs
#Generate isotropic Gaussian blobs for clustering.

from sklearn.datasets.samples_generator import make_blobs
(X1,y_train) = make_blobs(n_samples=300,n_features=2,centers=2,cluster_std=1.05,random_state=290)

X_train = np.c_[np.ones((X1.shape[0])),X1]#adding 1 as bias

plt.scatter(X_train[:,1],X_train[:,2],marker='o',c=y_train)
#plt.axis[-5,-10,-12,-1]
plt.show()

#splitting the dataset
#separating data for -1 and 1 classification
pos_X = []
neg_X = []
#Enumerate() method adds a counter to an iterable and returns it in a form of enumerate object.
for i,x in enumerate(y_train):
    if x==0:
        neg_X.append(X1[i])
    else:
        pos_X.append(X1[i])
        
#creating a data_dictionary
dd = {-1:np.array(neg_X),1:np.array(pos_X)}
weight =[]
weight
bais = []
bais
MAX = float('-inf')
MIN = float('+inf')
#np.amax: Return the maximum of an array or maximum along an axis.
for i in dd:
    if np.amax(dd[i])> MAX:
        MAX = np.amax(dd[i])
    if np.amin(dd[i])<MIN:
        MIN=np.amin(dd[i])

learn_rate = [MAX * 0.01]

vector = {}
transform = [[1,1],[-1,1],[-1,-1],[1,-1]]
step = 1
multiple = 2
w_final = MAX * 0.5

for lr in learn_rate:
    weight = np.array([w_final,w_final])
    optimal = False
    while not optimal:
        for bais in np.arange(-1*(MAX * step),MAX *step,lr*multiple):
            for trans in transform:
                w_t = weight * trans
                correct_classification = True
                for i in dd:
                    for j in dd[i]:
                        if i*(np.dot(w_t,j)+bais) < 1:
                            correct_classification = False
                if correct_classification:
                    vector[np.linalg.norm(w_t)] = [w_t,bais]
        
        if weight[0] < 0:
            optimal = True
        else:
            weight = weight - lr
        
    norms = sorted([n for n in vector])
    temp = vector[norms[0]]
    weight = temp[0]
    bais = temp[1]
    #w_final = weight[0]+lr *2
    #print(w_final)

y_pred = (np.sign(np.dot(np.array(X1),weight)+bais).astype(int))
y_train[y_train < 1] = -1    

#visualize
colors = {1:'r',-1:'b'}
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
#[[ax.scatter(X1[0],X1[1],c=colors[i]) for x in dd[i]] for i in dd]
        
plt.scatter(X1[:,0],X1[:,1],marker='o',c=y_train)


hyp_min = MIN*1.2
hyp_max = MAX*1

# (w.x+b) = 1
#positive support vector hyperplane
psv1 = hyperplane(hyp_min, weight, bais, 1)
psv2 = hyperplane(hyp_max, weight, bais, 1)
ax.plot([hyp_min,hyp_max],[psv1,psv2], 'k')
# (w.x+b) = -1
# negative support vector hyperplane
nsv1 = hyperplane(hyp_min, weight, bais, -1)
nsv2 = hyperplane(hyp_max, weight, bais, -1)
ax.plot([hyp_min,hyp_max],[nsv1,nsv2], 'k')
# (w.x+b) = 0
# Middle support vector hyperplane
db1 = hyperplane(hyp_min, weight, bais, 0)
db2 = hyperplane(hyp_max, weight, bais, 0)
ax.plot([hyp_min,hyp_max],[db1,db2], 'r--')
        
plt.axis([-5,10,-12,-1])
plt.show()
```
