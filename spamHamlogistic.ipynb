```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/home/kasat/datasets/spam_ham_data_set.csv',encoding='latin-1')
```

```python
del df['Unnamed: 2']
del df['Unnamed: 3']
del df['Unnamed: 4']

df = df.drop_duplicates()
df['v2'] = df['v2'].str.replace(r'[^\w\s]+', '')
df['v2'] = df['v2'].str.replace(r'[^A-Za-z0-9\s]','')
df['v1'] = df['v1'].replace('ham', 1)
df['v2'] = df['v2'].replace('spam', 0)
df.head()

# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
x = df['v2']
y = df['v1']
def create_unique_dict(text_arr):
    dictt = {}
    count = -1
    x = np.array(text_arr)
    text1 = (np.unique(x))

    for item in text1:
        data = (item.lower().strip().split(" "))
        for item1 in data:
            if item1 not in dictt:
                count = count +1
                dictt[item1] = count
    print((dictt))
    return dictt

def vectorize(text_arr,dictt):
    vectors  = []
    for sentence in text_arr:
        vector = [0] * len(dictt)
        words = (sentence.lower().strip().split(" "))
        for word in words:
            index = dictt[word]
            vector[index] += 1
        vectors.append(vector)
    return vectors

dictt = create_unique_dict(x)
x = np.array(vectorize(x, dictt))
y = y.replace('ham', 1)
y = y.replace('spam', 0)
X = df['v2']
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['v2'], y, test_size=0.3)
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def loss(h, y):
    return (-y * np.log(h) - (1 - y) * np.log(1 - h)).mean()
```

```python
def train(X,y, lr, num_iter):
    
    intercept = np.ones((X.shape[0], 1))
    X = np.concatenate((intercept, X), axis=1)
    
    # weights initialization
    theta = np.zeros(X.shape[1])

    for i in range(num_iter):
        z = np.dot(X, theta)
        h = sigmoid(z)
        gradient = np.dot(X.T, (h - y)) / y.size
        theta -= lr * gradient

        z = np.dot(X, theta)
        h = sigmoid(z)
        loss1 = loss(h, y)

        if( i % 100 == 0):
            print(f'loss: {loss1} \t')
    return theta
```

```python
def predict(X, theta, threshold):
    intercept = np.ones((X.shape[0], 1))
    X = np.concatenate((intercept, X), axis=1)
    prob = sigmoid(np.dot(X, theta))
    return prob >= threshold
```

```python
def evaluate(test_x, test_y, theta, threshold):
    y_predicted = predict(test_x, theta, threshold)
    correct = 0
    for i,y in enumerate(test_y):
        if y == 0:
            y = False
        else:
            y = True
        if y == y_predicted[i]:
            correct = correct + 1
    total = y_predicted.size

    return (correct/total)*100
```

```python
x = np.array(vectorize(X_train, dictt))
y = np.array(y_train)

test_x = np.array(vectorize(X_test, dictt))
test_y = np.array(y_test)
```

```python
lr=0.01
num_iter=1000
theta = train(x, y, lr, num_iter)
```

```python
print(evaluate(test_x,test_y, theta, 0.5))
```
